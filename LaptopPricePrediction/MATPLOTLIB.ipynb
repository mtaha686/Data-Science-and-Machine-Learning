```python
# Area Plot
x = np.linspace(0, 2 * np.pi, 100)
y = np.sin(x)
plt.figure(figsize=(8, 4))
plt.fill_between(x, y)
plt.title("Area Plot")
plt.show()
```

```python
# 3D Plot
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')

x3d = df['Ram']
y3d = df['SSD']
z3d = df['Weight']

ax.scatter(x3d, y3d, z3d)
ax.set_title("3D Scatter Plot")
plt.show()
```

```python
# Contour Plot
x_contour = np.linspace(-2, 2, 100)
y_contour = np.linspace(-2, 2, 100)
X, Y = np.meshgrid(x_contour, y_contour)
Z = X**2 + Y**2

plt.figure(figsize=(8, 6))
plt.contour(X, Y, Z, cmap='viridis')
plt.title("Contour Plot")
plt.show()
```

```python
# Polar Plot
theta = np.linspace(0, 2 * np.pi, 100)
r = 1 + np.sin(3 * theta)

plt.figure(figsize=(6, 6))
plt.polar(theta, r)
plt.title("Polar Plot")
plt.show()
```

```python
# Quiver Plot
x_quiver = np.linspace(-2, 2, 10)
y_quiver = np.linspace(-2, 2, 10)
X_quiver, Y_quiver = np.meshgrid(x_quiver, y_quiver)
U = -1 - X_quiver**2 + Y_quiver
V = 1 + X_quiver - Y_quiver**2

plt.figure(figsize=(8, 6))
plt.quiver(X_quiver, Y_quiver, U, V, scale=20)
plt.title("Quiver Plot")
plt.show()
```

```python
# Error Bar Plot
x_error = np.arange(0, 5, 1)
y_error = x_error**2
error = np.random.rand(5)

plt.figure(figsize=(8, 4))
plt.errorbar(x_error, y_error, yerr=error, fmt='o', capsize=5)
plt.title("Error Bar Plot")
plt.show()
```

```python
# Hexbin Plot
x_hexbin = np.random.randn(1000)
y_hexbin = np.random.randn(1000)

plt.figure(figsize=(8, 6))
plt.hexbin(x_hexbin, y_hexbin, gridsize=25, cmap='viridis')
plt.colorbar()
plt.title("Hexbin Plot")
plt.show()
```

```python
labels = ['A', 'B', 'C']
sizes = [15, 30, 45]

plt.figure(figsize=(6, 6))
plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.title("Pie Chart")
plt.show()
```
